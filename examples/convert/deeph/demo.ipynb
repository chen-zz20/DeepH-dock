# DeepH Interface

## Upgrade

User can convert data from DeepH-E3/DeepH-2 formats to the updated DeepH-pack specification.

```bash
dock convert deeph upgrade legacy.bak updated -t 0 -p 2
```

## Downgrade
User can convert data from DeepH-pack format back to DeepH-E3/DeepH-2 formats.

```bash
dock convert deeph downgrade updated.bak legacy -t 0 -p 2
```

## Standardize
User can standardize DeepH Hamiltonian to eliminate the mu gauge.

```bash
dock convert deeph standardize standardize --overwrite -t 0 -p 2
```

## Minus core

User can remove or add back the single atomic Hamiltonian, which significantly reduces the range of Hamiltonian values.

```bash
for d1 in $(ls updated); do
  dock convert deeph minus-core updated/$d1 minus_core/$d1 single_atoms.bak -t -1 -p 1
done
```
