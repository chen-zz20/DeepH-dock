# Hopcp Interface

## Install PETSc

```bash
dock convert hopcp how-to-install-petsc
```

## Begin with Terminal Tool

User can use the terminal tool to convert DeepH format to Hopcp format.
```bash
dock convert hopcp from-deeph ./deeph_data /tmp/petsc_data -t 0 -p 1
```

## Advanced Usage

````python
from deepx_dock.convert.hopcp.translate_deeph_to_petsc import DeepHtoPETScTranslator

deeph_dir = "./deeph_data"
petsc_dir = "/tmp/petsc_data"
ignore_s = False
ignore_h = False
parallel_num = 1
tier_num = 0

transfer = DeepHtoPETScTranslator(
    deeph_dir=deeph_dir,
    petsc_dir=petsc_dir,
    export_S=not ignore_s,
    export_H=not ignore_h,
    n_jobs=parallel_num,
    n_tier=tier_num,
)
transfer.transfer_all_deeph_to_petsc()
````
